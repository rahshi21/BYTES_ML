**One Hot Encoding using Sci-kit Learn Library**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse=False)
df_encoded = pd.DataFrame(onehot.fit_transform(df[['Loan_ID','Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Credit_History', 'Property_Area', 'Loan_Status']]),
                          columns = onehot.get_feature_names_out(['Loan_ID','Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Credit_History', 'Property_Area', 'Loan_Status']))
```

**manual mapping**

```python
df['Gender'] = df['Gender'].map({"Male" : 1, "Female" : 0})
df['Married'] = df['Married'].map({"Yes" : 1, "No" : 0})
df['Education'] = df['Education'].map({"Graduate" : 1, "Not Graduate" : 0})
df['Dependents'].replace('3+', 3, inplace=True)
df['Self_Employed'] = df['Self_Employed'].map({"Yes" : 1, "No" : 0})
df['Property_Area'] = df['Property_Area'].map({"Semiurban" : 1, "Urban" : 2, "Rural" : 3})
df['Loan_Status'] = df['Loan_Status'].map({"Y" : 1, "N" : 0})
```

```python
df.isnull().sum()
```

```python
df
```

```python
import numpy as np
rev_null = ['Gender','Married','Dependents','Self_Employed','Credit_History','LoanAmount','Loan_Amount_Term']
df[rev_null] = df[rev_null].replace({np.nan:df['Gender'].mode(),
                                     np.nan:df['Married'].mode(),
                                     np.nan:df['Dependents'].mode(),
                                     np.nan:df['Self_Employed'].mode(),
                                     np.nan:df['Credit_History'].mode(),
                                     np.nan:df['LoanAmount'].mean(),
                                     np.nan:df['Loan_Amount_Term'].mean(),})
```

**train test**

```python
X = df.drop(columns=['Loan_ID','Loan_Status']).values
y = df['Loan_Status'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train
```

**SCALING**

min max scaling -> convert values to 0's and 1's

standard scaling -> convert values to the given range
                   
standard      scaling           -> values in ranges -1 to 1

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train = sta.fit_transform(X_train)
X_test = sta.transform(X_test)
```

**MODEL BUILDING**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc = RandomForestClassifier(criterion='entropy',random_state= 42)
rfc.fit(X_train, y_train)
```

```python
#Evaluating on traning set
y_pred_train = rfc.predict(X_train)
print(classification_report(y_train, y_pred_train))
```
